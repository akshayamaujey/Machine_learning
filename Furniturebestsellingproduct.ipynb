Problem Statement: We have the following datasets and column:
 1. customers - customer_id ,names
 2. products=product_id,product_name, price ,category
 3. ratings=product_id , customer_id
  4.orders= order_id,customer_id, product_id,order_date,quantity                               
  Here i wish to analyze the data and find out top performing products in terms of revenue as well as in terms of how many of  units have been sold and identify the top client for the last month.

```python
import pandas as pd

# Sample code to load your datasets - replace with your actual data loading code
customers = pd.read_csv('/content/customers.csv')
products = pd.read_csv('/content/products.csv')
ratings = pd.read_csv('/content/ratinggpt.csv')
orders = pd.read_csv('/content/orders.csv')


# Step 1: Merge DataFrames
orders_products = pd.merge(orders, products, on='product_id')

orders_products
```

```python
# Step 2: Calculate Revenue for Each Product
orders_products['revenue'] = orders_products['quantity'] * orders_products['price']
orders_customers_products = pd.merge(orders_products, customers, on='customer_id')
orders_customers_products
```

```python
# Step 3: Count Units Sold for Each Product
total_revenue_per_product = orders_products.groupby('product_id')['revenue'].sum()
units_sold_per_product = orders_products.groupby('product_id')['quantity'].sum()
total_revenue_per_product,total_revenue_per_product
```

```python
orders_customers_products['order_date'] = pd.to_datetime(orders_customers_products['order_date'])
```

```python
# Find the last month from the data
last_month = orders_customers_products['order_date'].dt.to_period('M').max()
last_month_orders = orders_customers_products[orders_customers_products['order_date'].dt.to_period('M') == last_month]
last_month,last_month_orders
```

```python
# Calculate total purchase for each customer in the last month and find the top customer
top_client_last_month = last_month_orders.groupby('customer_id')['revenue'].sum().idxmax()
top_client_last_month
```

```python
# Display the results
print("Total Revenue Per Product:\n", total_revenue_per_product)
print("\nUnits Sold Per Product:\n", units_sold_per_product)
print("\nTop Client for Last Month (Customer ID):", top_client_last_month)
```
